```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import re
import json
import urllib.request
from selenium import webdriver
import time
```

```python
file_path = '/2nd_project/ingredient.json'
def toJson(recipe_dict):
    with open(file_path, 'w', encoding='utf-8') as file:
        json.dump(recipe_dict, file, ensure_ascii=False, indent='\t')
```

```python
# ingredient_list = []
ingredient_lists = []
num_id = 1
for i in range(1041):
    driver = webdriver.Chrome('/learn/TIL/4th/파이썬/data_crawling/data/chromedriver')
    # 웹드라이버 실행 경로 chromedriver는 폴더가 아니라 파일명입니다.
    driver.get('https://www.10000recipe.com/recipe/6881099');
    time.sleep(1)
    driver.execute_script(f'viewMaterial({i})')
    driver.implicitly_wait(5)
    time.sleep(2)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    res = soup.find('div','ingredient_wrap')
    if not(res):
        continue
    driver.quit()

    #name
    res = soup.find('div','ingredient_tit')
    name = res.get_text()

    #info
    res = soup.find('div','ingredient_info')
    info = []
    c1 = []
    c2 = []
    for m in res.find_all('th'):
        h1 = m.get_text().replace('\n','').replace(' ','')
        c1.append(h1)
    for n in res.find_all('td'):
        h2 = n.get_text().replace('\n','').replace('  ','')
        c2.append(h2)

    for i in range(len(c1)):
        info.append(c1[i]+': '+c2[i])

    #trim
    trim = []
    eff = '효능'
    for n in soup.find_all('dl','ingredient_cont'):    
        for m in n.find_all('dt'):
            h1 = m.get_text().replace('\n','').replace(' ','')
        for j in n.find_all('dd'):
            h2 = j.get_text().replace('\n','').replace('  ','')
        if h1 == eff:
                continue
        trim.append(h1+': '+h2)
        
    ingredient_dict = {"id":num_id,
                "name":name,
                "info":info,
                "trim":trim}
    ingredient_list.append(ingredient_dict)
    num_id += 1
    
toJson(ingredient_lists)
```
