```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

train = pd.read_csv('train.csv',index_col='PassengerId')
# print(train.shape)
# train.head(4)
```

```python
# 1. 타이타닉의 train 데이터에서 1) 전체 생존률과 2) 생존자의 총 인원수, 사망자의 총 인원수를 출력해주세요.

survived_rate = train.loc[:,'Survived'].mean() * 100
alive = train.loc[train.Survived == 0,'Survived'].count()
dead = train.loc[train.Survived == 1,'Survived'].count()
print(survived_rate)
print(alive,dead)
```

```python
# **** <문제3> ****
# Survived 컬럼의 상위 5개의 값을 출력합니다.
# 결과값은 0과 1이 나오는데, Survived 컬럼에 대한 사전 설명(가령 0이 어떤 값을 나타내는지, 1이 어떤 값을 나타내는지)
# 을 듣지 않으면 이 값이 어떠한 의미를 가지는지 직관적으로 이해하기 어렵습니다.

train.loc[:,'Survived'].head()
```

```python
# Survived(humanized)라는 새로운 컬럼 만들기
#이 컬럼에는 0(사망), 1(생존)이 아닌, Perish(사망), Survived(생존). 
#최종적으로는 다음의 결과가 나와야 합니다.

# temp = train['Survived']
# print(type(temp))
# train['Survived(humanized)'] = temp

train['Survived(humanized)'] = np.where(train['Survived'] == 1,'Survived',train['Survived(humanized)'])
train.head()
```

```python
train['Survived(humanized)'].value_counts()
```

```python
pd.pivot_table(data=train, index='Pclass',values='Survived')
```

```python
# temp = train['Pclass']
# train['Pclass(humanized)'] = temp
train['Pclass(humanized)'] = np.where(train['Pclass'] == 1,'First Class',train['Pclass(humanized)'])
train['Pclass(humanized)'] = np.where(train['Pclass'] == 2,'Business',train['Pclass(humanized)'])
train['Pclass(humanized)'] = np.where(train['Pclass'] == 3,'Economy',train['Pclass(humanized)'])
train.loc[:,['Pclass','Pclass(humanized)']].head()
```

```python
pd.pivot_table(data=train, index='Pclass(humanized)',values='Survived')
```

```python
temp = train['Embarked']
train['Embarked(humanized)'] = temp
train['Embarked(humanized)'] = np.where(train['Embarked'] == 'S','Southampton',train['Embarked(humanized)'])
train['Embarked(humanized)'] = np.where(train['Embarked'] == 'Q','Queenstown',train['Embarked(humanized)'])
train['Embarked(humanized)'] = np.where(train['Embarked'] == 'C','Cherbourg',train['Embarked(humanized)'])

pd.pivot_table(data=train, index='Embarked(humanized)',values='Survived')
```

```python
train['Pclass(FirstClass)'] = False
train['Pclass(Business)'] = False
train['Pclass(Economy)'] = False
train['Pclass(FirstClass)'] = np.where(train['Pclass'] == 1,True,train['Pclass(FirstClass)'])
train['Pclass(Business)'] = np.where(train['Pclass'] == 2,True,train['Pclass(Business)'])
train['Pclass(Economy)'] = np.where(train['Pclass'] == 3,True,train['Pclass(Economy)'])
train.loc[:,['Pclass','Pclass(humanized)','Pclass(FirstClass)','Pclass(Business)','Pclass(Economy)']].head(10)
```

```python
train['Embarked(Cherbourg)'] = False
train['Embarked(Southampton)'] = False
train['Embarked(Queenstown)'] = False
train['Embarked(Cherbourg)'] = np.where(train['Embarked'] == 'C',True,train['Embarked(Cherbourg)'])
train['Embarked(Southampton)'] = np.where(train['Embarked'] == 'S',True,train['Embarked(Southampton)'])
train['Embarked(Queenstown)'] = np.where(train['Embarked'] == 'Q',True,train['Embarked(Queenstown)'])
train.loc[:,['Embarked','Embarked(humanized)','Embarked(Cherbourg)','Embarked(Southampton)','Embarked(Queenstown)']].head(10)
```
